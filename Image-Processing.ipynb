```python
# Resizing the image using Pillow library
from PIL import Image
```

```python
img = Image.open('dog.jpeg')
img_resized = img.resize((200, 200))
```

```python
img_resized.save('dog_resized.jpeg')
```

```python
img_resized_r = mpimg.imread('dog_resized.jpeg')
plt.imshow(img_resized_r)
print(img_resized_r.shape)
```

```python
# Converting RGB images to Grayscale image using Opencv
```

```python
import cv2 # importing opencv library
```

```python
img = cv2.imread('dog_resized.jpeg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(grayscale_image, cmap="gray")
plt.show
```

```python
cv2.imwrite('dog-grayscaled.jpeg', grayscale_image)
```
