```python
original_folder = 'image_resized'
resized_folder = 'image_resized_3'

for i in range(2014):
    
    file_name = os.listdir(original_folder)[i]
    img_path = original_folder+'/'+file_name
    
    img = Image.open(img_path)
    img = img.resize((224, 224))
    img = img.convert('RGB')
    
    newImgPath = resized_folder+'/'+file_name
    img.save(newImgPath)
```

```python
import os
# creating labels for resized images
file_names = os.listdir('image_resized_3')
labels = []

for i in range(2014):
    
    file_name = file_names[i]
    label = file_name[0:3]
    
    if label == 'dog':
        labels.append(1) # Dog
    else:
        labels.append(0) # Cat
```

```python
#Converting all the resized images to numpy array
import cv2
import glob
```

```python
image_directory = 'image_resized_3/'
image_extension = ['png', 'jpg']

files = []

[files.extend(glob.glob(image_directory + '*.' + e)) for e in image_extension]
dog_cat_images = np.asarray([cv2.imread(file) for file in files])
```

```python
X = dog_cat_images
Y = np.asarray(labels)
```

```python
# Train Test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape) # 8000 training images, 2000 test images
```

```python
# scaling the data
X_train_scaled = X_train/255

X_test_scaled = X_test/255
```

```python
print(X_train_scaled)
```

```python
import tensorflow as tf
import tensorflow_hub as hub
```

```python
mobilenet_model = 'https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4'

pretrained_model = hub.KerasLayer(mobilenet_model, input_shape = (224,224, 3), trainable=False)
```

```python
num_of_classes = 2
model = tf.keras.Sequential([
    pretrained_model, 
    tf.keras.layers.Dense(num_of_classes)
])

model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics = ['acc']
)
```

```python
model.fit(X_train_scaled, Y_train, epochs=5)
```

```python
score, acc = model.evaluate(X_test_scaled, Y_test)
print('Test loss =', score)
print('Test accuracy =', acc)
```

```python
# Predictive system
input_image_path = input('Path of the image to be predicted: ')
input_image = cv2.imread(input_image_path)

cv2.imshow('',input_image)
cv2.waitkey(0)

input_image_resize = cv2.resize(input_image, (224, 224))

input_image_scaled = input_image_resize/255

image_reshaped = np.reshape(input_image_scaled, [1, 224, 224, 3])

input_prediction = model.predict(image_reshaped)

input_pred_label = np.argmax(input_prediction)

if input_pred_label == 0:
    print('The image represents a Cat')
else:
    print('The image represents a Dog')
```
